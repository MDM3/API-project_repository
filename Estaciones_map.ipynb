```python
import folium

# Crea el mapa centrado en Madrid
m = folium.Map(location=[40.416775, -3.703790], zoom_start=10)

# Lista de estaciones
estaciones = [
    {
        "NOMBRE": "MADRID, RETIRO",
        "INDICATIVO": "3195",
        "PROVINCIA": "MADRID",
        "ALTITUD": 667.0,
        "COORDENADAS": [-3.67806112, 40.411950352000019]  
    },
    {
        "NOMBRE": "MADRID, CUATRO VIENTOS",
        "INDICATIVO": "3196",
        "PROVINCIA": "MADRID",
        "ALTITUD": 690.0,
        "COORDENADAS": [-3.786105280000015, 40.375554352999984]
    },
    {
        "NOMBRE": "MADRID, BARAJAS",
        "INDICATIVO": "3129",
        "PROVINCIA": "MADRID",
        "ALTITUD": 609.0,
        "COORDENADAS": [-3.555558222, 40.466667176999977]
    },
    {
        "NOMBRE": "MADRID, EL GOLOSO",
        "INDICATIVO": "3126Y",
        "PROVINCIA": "MADRID",
        "ALTITUD": 650.0,
        "COORDENADAS": [-3.709128, 40.552986]
    }
]

# Agrego los marcadores para las estaciones
for estacion in estaciones:
    folium.Marker(
        location=[estacion["COORDENADAS"][1], estacion["COORDENADAS"][0]],  
        popup=folium.Popup(
            f"<b>Nombre:</b> {estacion['NOMBRE']}<br>"
            f"<b>Indicativo:</b> {estacion['INDICATIVO']}<br>"
            f"<b>Provincia:</b> {estacion['PROVINCIA']}<br>"
            f"<b>Altitud:</b> {estacion['ALTITUD']} m",
            max_width=300
        )
    ).add_to(m)

# Con esto agrego las diferentes capas del visor
folium.TileLayer('OpenStreetMap').add_to(m)
folium.TileLayer('Stamen Terrain').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('cartodb positron').add_to(m)
folium.TileLayer('cartodb dark_matter').add_to(m)

# Asigno el visor 'satelite' por defecto
folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='&copy; <a href="https://www.esri.com/">Esri</a>',
    name='Esri Satellite',
    overlay=False,
    control=True
).add_to(m)


folium.LayerControl().add_to(m)

# Exporto el mapa en un archivo HTML
m.save("mapa.html")

# Muestro el mapa
m
```
