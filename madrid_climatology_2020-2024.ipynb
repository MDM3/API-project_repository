```python
import pandas as pd
import glob  # Libreria para buscar archivos en las carpetas
import os
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>2. Importación de los archivos</p>
</div>


```python
path = r'C:\Users\Darkos\Desktop\Labs\WEEK_3_First_porject_'
path1 = r"C:\Users\Darkos\Desktop\Labs\WEEK_3_First_porject_\master\madrid_retiro_1980"
dataframes = []
df_retiro_1980 = []

''''Se itera en las carpetas para incluis los csv's en las listas creadas'''

for archive in glob.glob(os.path.join(path, '*.csv')):
    df = pd.read_csv(archive)
    dataframes.append(df)

for element in glob.glob(os.path.join(path1, '*.csv')): 
    df_retiro = pd.read_csv(element)
    df_retiro_1980.append(df_retiro)
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>3. Concatenación de los dataframes.</p>
      <ul>
    <li style='font-size:22px'>df_concatenated</li>
    <li style='font-size:22px'>df_historic</li>
    </ul>
</div>

```python
'''Concatenacion de ambos dataframes'''
df_concatenated = pd.concat(dataframes, ignore_index=True)
df_historic = pd.concat(df_retiro_1980, ignore_index=True)
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>4. Supresión de columnas</p>
</div>

```python
df_concatenated = df_concatenated.drop(['sol','presMax','horaPresMax', 'presMin','horaPresMin','hrMin','horaHrMin','horaracha','hrMedia','hrMax','horaHrMax','racha'], axis=1)
df_historic = df_historic.drop(['n_cub','hr','n_gra','n_fog','inso','q_max','nw_55','q_mar','q_med','ta_max','ts_min','ta_min','nt_30','nv_0050','n_des','w_racha','np_100','n_nub','p_sol','nw_91','np_001','w_rec','e', 'np_300','nv_1000','evap','p_mes','n_llu','n_tor','w_med','nt_00','ti_max','n_nie','nv_0100','q_min','np_010','ts_20','ts_10','ts_50','Unnamed: 0'],axis=1)
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>5. Funcion grados </p>
</div>

```python
'''Funcion que pasa los grados de la direccion del viento y devuelve un string de las orientaciones'''
def dir(grades):
    if 1 <= grades <= 12:
        return "N"
    elif 13 <= grades <= 24:
        return "NE"
    elif 25 <= grades <= 36:
        return "E"
    elif 37 <= grades <= 48:
        return "SE"
    elif 49 <= grades <= 60:
        return "S"
    elif 61 <= grades <= 72:
        return "SW"
    elif 73 <= grades <= 84:
        return "W"
    elif 85 <= grades <= 99:
        return "NW"
    
df_concatenated['dir'] = df_concatenated['dir'].apply(dir)
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>6. Estaciones de Madrid. Limpieza de datos y manipulación de datos. </p>

```python
df_concatenated['dir'] = df_concatenated['dir'].ffill()
df_concatenated['tmin'] = df_concatenated['tmin'].ffill()
df_concatenated['tmax'] = df_concatenated['tmax'].ffill()
df_concatenated['prec'] = df_concatenated['prec'].ffill()
df_concatenated['tmed'] = df_concatenated['tmed'].ffill()
```

```python
df_concatenated['horatmin'] = pd.to_datetime(df_concatenated['horatmin'],errors='coerce').dt.time
df_concatenated['horatmax'] = pd.to_datetime(df_concatenated['horatmax'],errors='coerce').dt.time
df_concatenated['tmed'] = df_concatenated['tmed'].str.replace(',', '.').astype(float)
df_concatenated['tmax'] = df_concatenated['tmax'].str.replace(',', '.').astype(float)
df_concatenated['tmin'] = df_concatenated['tmin'].str.replace(',', '.').astype(float)
df_concatenated['velmedia'] = df_concatenated['velmedia'].str.replace(',', '.').astype(float)
```

    <p>6.1 Mapeo columna fecha </p>
</div>

```python
df_concatenated['año'] = df_concatenated['fecha'].str.split('-').str[0]
df_concatenated['mes'] = df_concatenated['fecha'].str.split('-').str[1]
df_concatenated['dia'] = df_concatenated['fecha'].str.split('-').str[1]

meses_df_concatened = {
    '01': 'Enero',
    '02': 'Febrero',
    '03': 'Marzo',
    '04': 'Abril',
    '05': 'Mayo',
    '06': 'Junio',
    '07': 'Julio',
    '08': 'Agosto',
    '09': 'Septiembre',
    '10': 'Octubre',
    '11': 'Noviembre',
    '12': 'Diciembre'
}

df_concatenated['mes_nombre'] = df_concatenated['mes'].map(meses_df_concatened)
```

<div style="font-family: 'Roboto', sans-serif; font-size: 25px; color: white;">
    <p>7. Estación El Retiro. Historico 1980-2023. Limpieza de datos y manipulación de datos. </p>

</div>

```python
df_historic['año'] = df_historic['fecha'].str.split('-').str[0]
df_historic['mes'] = df_historic['fecha'].str.split('-').str[1]

meses = {
    '1': 'Enero',
    '2': 'Febrero',
    '3': 'Marzo',
    '4': 'Abril',
    '5': 'Mayo',
    '6': 'Junio',
    '7': 'Julio',
    '8': 'Agosto',
    '9': 'Septiembre',
    '10': 'Octubre',
    '11': 'Noviembre',
    '12': 'Diciembre'
}

df_historic['nombre_mes'] = df_historic['mes'].map(meses) #mapeo de para la columna nombre-mes
df_historic = df_historic.drop(columns='fecha') #dropeo de la columna fecha, ya no la quiero
df_historic = df_historic[df_historic['mes'] != '13'] #eliminar registros con el mes = 13

historic_columns_order = ['año','mes','nombre_mes', 'tm_min', 'tm_max', 'tm_mes', 'p_max','indicativo']
df_historic = df_historic[historic_columns_order]

df_historic = df_historic.dropna(subset=['tm_min'])
df_historic['mes'] = df_historic['mes'].astype(int)

df_historic = df_historic.sort_values(by=['año', 'mes'], ascending=[True, True])
```
